#### Data Preprocessing an visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder

#loading data
df = pd.read_csv("titanic-passengers.csv", sep=";")
df
```

```python
df.head(20)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum() #sum of null values by column
```

```python
df.isnull().sum().sum()  # sum of all missing values
```

```python
mean_age = round(df["Age"].mean()) #calculate age mean
mean_age
```

```python
df["Age"].fillna(mean_age,inplace=True) #replace null values by mean of ages
df["Cabin"].fillna("FFR", inplace=True)
```

```python
df.isnull().sum() #sum of null values by column
```

```python
df["Cabin"].value_counts() 
```

```python
#replace NaN values by the most common value
df["Cabin"].fillna("G6", inplace=True)
```

```python
df.isnull().sum()
```

```python
df["Embarked"].value_counts()
```

```python
df["Embarked"].fillna("S", inplace= True)
```

```python
df.isnull().sum()
```

```python
#delete useless columns
df
df.drop(["PassengerId","Ticket", "Name","Parch"],inplace=True, axis=1)
df
```

```python
#codification of categorical Data ==> using label encoding
label_encoder = LabelEncoder()
df_encoded = df
df_encoded["Survived"]=label_encoder.fit_transform(df_encoded["Survived"])
#df_encoded["Sex"]=label_encoder.fit_transform(df_encoded["Sex"])
df_encoded["Cabin"]=label_encoder.fit_transform(df_encoded["Cabin"])
df_encoded["Embarked"]=label_encoder.fit_transform(df_encoded["Embarked"])
df_encoded
```

```python
#scaling
from sklearn.preprocessing import StandardScaler
df_scaled = df_encoded
df_scaled["Fare"] = StandardScaler().fit_transform(df_encoded["Fare"].values.reshape(-1, 1))
#df_scaled["Fare"] = StandardScaler().fit(df_scaled)
df_scaled
```

```python
#distribution => boxPlot
df_dist = df_scaled
sns.boxplot(x='Age', y="Sex",data = df_dist,hue="Survived",saturation=1, palette="Set2",linewidth=2.5) 
```

```python
sns.distplot(df_dist['Age'],bins=10,hist=True,kde=True,color='red')
```

```python
sns.countplot(x='Sex',data=df_dist)
plt.xticks(rotation=45)
```

```python

sns.countplot(x='Survived',data=df_dist)
plt.xticks(rotation=45)
```
